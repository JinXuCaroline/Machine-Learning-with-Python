```python
# read excel
import pandas as pd

xls_file = pd.ExcelFile("News_Category_Analysis.xlsx")
xls_file.sheet_names
unigram_file = xls_file.parse("gram1")
unigram_file.columns = ['keyword', 'frequency']
unigram_file['keyword'].str.replace(',', '')
#bigram_file = pd.read_excel("News_Category_Analysis.xlsx", )
unigram_file_sort = unigram_file.sort_values(ascending = False, by = 'frequency')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
unigram_file_head = unigram_file_sort.head(20)
ax = sns.barplot(x='frequency', y='keyword', data=unigram_file_head)
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
d = {}
for k, f in unigram_file_sort.values:
    d[k] = f
wordcloud = WordCloud(background_color = 'white', max_font_size = 40, max_words = 100)
wordcloud.generate_from_frequencies(frequencies = d)
plt.figure()
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.show()  
```
